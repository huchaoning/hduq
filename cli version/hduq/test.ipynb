```python
import numpy as np
from itertools import product

A = np.array([[HG(0, 0), HG(0, 1)],
              [HG(1, 0), HG(1, 1)]])
B = np.array([[HG(0, 0), HG(0, 1)],
              [HG(1, 0), HG(1, 1)]])

# 拉平成一维数组
a_flat = A.flatten()
b_flat = B.flatten()

# 所有 (a, b) 组合
combinations = list(product(a_flat, b_flat))

# 输出
for a, b in combinations:
    print(f"(a, b) = ({a}, {b})")
```
